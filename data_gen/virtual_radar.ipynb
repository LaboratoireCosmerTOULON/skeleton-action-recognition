```python
from skimage.transform import resize
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams['image.cmap'] = 'jet'

import sys
sys.path.extend(['../'])
from data_gen.gen_spectrogram_data import *
```

```python
fname = '../data/nturgbd_raw/nturgb+d_skeletons/S001C002P004R002A057.skeleton'
data, edges = preprocess_ntu(fname)
spectrogram, range_map = synthetic_spectrogram_ntu(data, edges, 
                                                   num_pad_frames=500, 
                                                   radar_lambda=4e-4)

print(spectrogram.shape)
im = plt.imshow(resize(spectrogram, (256, 512))) 
im.set_clim(-35, 0)
plt.show()

im = plt.imshow(resize(range_map, (256, 512))) 
plt.show()
```
