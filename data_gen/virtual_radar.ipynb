```python
def read_xyz_rgb(file, max_body=4, num_joint=17):
    data = np.load(file)
    if data.shape[0] == 1:
        data = np.concatenate((data, np.zeros_like(data)), axis=0)
    data = data.transpose(3, 1, 2, 0)
    return data
```
